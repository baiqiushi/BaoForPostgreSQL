```python
import numpy as np
import string
```

```python
def read_bao_data(fp):
    with open(fp) as f:
        data = f.read().split("\n")[2:]

    training_times = []
    for idx in range(len(data)):
        if data[idx].strip().startswith("Initial input channels"):
            prev_line = data[idx-1].split(" ")
            if prev_line[0] == "Retry":
                continue
            training_times.append(float(prev_line[2]))


    training_times = np.array(training_times)

    data = [x.split(" ") for x in data if len(x) > 1 and (x[0] in string.digits or x[0] == "x")]
    data = [(x[0], x[1], float(x[2]), x[3], float(x[4])) for x in data]
    bao_data = data

    bao_times = np.array([x[4] for x in bao_data])

    return bao_times

bao_times = read_bao_data("bao_run.txt")
print("Running times of queries using Bao:")
#print(bao_times)
for bt in bao_times:
    print(bt)


# Read PG resutls
with open("pg_run.txt") as f:
    data = f.read().split("\n")[2:]
data = [x.split(" ") for x in data if len(x) > 1 and (x[0] in string.digits or x[0] == "x")]
data = [(x[0], x[1], float(x[2]), x[3], float(x[4])) for x in data]
pg_data = data
pg_times = np.array([x[4] for x in pg_data])
print("Running times of queries using Postgres:")
#print(pg_times)
for pt in pg_times:
    print(pt)

queries = np.array([x[3] for x in pg_data])
print("Queries:")
#print(queries)
for q in queries:
    print(q)
```

```python
print("Bao VQN = " + str(np.count_nonzero(bao_times < 0.4)))
print("Bao VQP = " + str(np.count_nonzero(bao_times < 0.4) / 100.0))
print("PG VQN = " + str(np.count_nonzero(pg_times < 0.4)))
print("PG VQP = " + str(np.count_nonzero(pg_times < 0.4) / 100.0))
```
